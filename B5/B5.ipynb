```python
Соединим всё, что мы делали, и очистим файл log.csv. А именно:

прочитаем файл в переменную log;
добавим названия колонок user_id, time, bet, win;
удалим значения user_id с ошибками;
оставим в user_id только значения идентификатора;
уберём начальную скобку из поля time.
```

```python
log = pd.read_csv("log.csv",header=None) 
log.columns = ['user_id','time','bet','win']

def rep(s):
    if s == "#error":
        return ""
    else:
        return s[s.find('user'):]
    
log.user_id = log.user_id.apply(rep)
log.time = log.time.str.replace('[', '')
```

# Избавляемся от ошибок/заменяем значения

```python
import pandas as pd
log=pd.read_csv('log.csv', header=None)
log.columns=['user_id','time','bet','win']
log = log.query('user_id != "#error"')

def user_id_correct(s):
      s=str(s).replace('Запись пользователя № - ','')
      return s

log.user_id=log.user_id.apply(user_id_correct)
log.time=log.time.str.replace('[','')
```
